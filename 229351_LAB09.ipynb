### ปฏิบัติการครั้งที่ 7
1. สร้างโมเดลเพื่อการทำนายดังนี้
1.1 แบ่งข้อมูลออกเป็น 2 ส่วน
   - training set: วันที่ 1996-01-01 ถึง 2009-12-01
   - test set: วันที่ 2010-01-01 ถึง 2012-03-01  
1.2 แยกส่วนประกอบ $y_t=S_t+T_t+R_t$ บน training set
1.3 สร้าง time series ชุดใหม่ที่แสดงถึงทำนายค่าบน test set โดยนำค่า $T_t+R_t$ ของวันล่าสุดใน training set ที่มีค่า $T_t$ มาบวกกับแต่ละค่าใน $S_t$ จากวันที่ วันที่ 2010-01-01 ถึง 2012-03-01  
3. คำนวณ RMSE โดยใช้ฟังก์ชัน `rmse` ข้างล่าง
4. แสดงแผนภาพข้อมูล elecequip และค่าทำนายที่ได้

```python
# 1.1
train_data = data.loc['1996-01-01':'2009-12-01']
test_data = data.loc['2010-01-01':'2012-03-01']

print("Training data shape:", train_data.shape)
print("Test data shape:", test_data.shape)
```

```python
# 1.2
result_train_add = seasonal_decompose(train_data['value'], model='additive', period=12)

last_trend_train = result_train_add.trend.dropna().iloc[-1]
last_resid_train = result_train_add.resid.dropna().iloc[-1]
seasonal_pattern = result_train_add.seasonal[-12:]
num_periods_test = len(test_data)
num_repeats = num_periods_test // 12 + 1 # Cal  how many times to repeat the pattern
repeated_seasonal = np.tile(seasonal_pattern, num_repeats)[:num_periods_test]
# Create a pandas Series
seasonal_test = pd.Series(repeated_seasonal, index=test_data.index)

# forecast the test set
forecast = last_trend_train + last_resid_train + seasonal_test
display(forecast)
```

```python
def rmse(y_true,y_pred):
  # Inputs
  # y_true: actual value, y_pred: predicted values
  # Output: RMSE between y_true and y_pred
  return np.sqrt(np.mean(np.square(y_true-y_pred)))
```

```python
# Cal the RMSE
rmse_value = rmse(test_data['value'], forecast)

print("RMSE:", rmse_value)
```
